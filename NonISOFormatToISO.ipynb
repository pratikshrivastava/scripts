```python
####################### 
# NON ISO Date
####################### 

import re
from datetime import datetime
# Function definition is here
def non_iso_to_iso( str, mm_dd ):
    ##This prints a passed string into this function"
    
    months={'JANUARY':'01','FEBRUARY':'02','MARCH':'03','APRIL':'04', 'MAY':'05', 'JUNE':'06',
            'JULY':'07', 'AUGUST':'08','SEPTEMBER':'09','OCTIBER':'10', 'NOVEMBER':'11', 'DECEMBER':'12'}
    
    ## trim the whitespaces in the end and newline character if any. 
    str=str.strip()
    str=str.strip("\n") 
       
    ## check whether the separator is "/" or "."
    if(str[1] == "/" or str[2] == "/"):
        #print(str)
        m=str[:str.find("/")]
        if(len(m) <2):
            m="0"+m
        #print(m)
        d=str[str.find("/")+1:str.find("/",str.find("/")+1)]
        #print(d)
        if(len(d) <2):
            d="0"+d
        y=str[str.find(d)+len(d)+1:]
        #print(y)
           
    elif(str[1] == "." or str[2] == "."):
    #    print(str)
        m=str[:str.find(".")]
        #print(str.find("."))
        if(len(m) <2):
            m="0"+m
        #print(m)
        d=str[str.find(".")+1:str.find(".",str.find(".")+1)]
        y=str[str.find(d)+len(d)+1:]
        if(len(d) <2):
            d="0"+d
        #print(y)
    else:
        #print(str)
        m=str[:str.find(" ")]
        m=months[(m.strip()).upper()]
        d=str[str.find(" ")+1:str.find(" ",str.find(" ")+1)]      
        d=d[:len(d)-1]
        y=str[str.find(d)+len(d)+1:]
        if(len(d) <2):
            d="0"+d
    if(mm_dd == 0 ):
        iso_dt=y+"-"+ m +"-"+ d
    elif(mm_dd == 1 ):
        iso_dt=y+"-"+ d +"-"+ m
    
    iso_dt=iso_dt.strip()+"\n"
    
    return iso_dt

####################### Part A #####################################
## For Finding dates having MM/DD/YYYY or MM.DD.YYYY format 

## find the lines from the files. 
## first open the file and loop through the lines 

print ("All the dates which matches in the format of MM/DD/YYYY or MM.DD.YYYY")
print ("---------------------------------------------------------------------")

fout=open('iso_date.txt','w')
for line in (open('date.txt','r+')):
    
    ## first select the month in 2 digit followed by "/ or ." used a 0(1-9) & 1(0-2) 
    ## second select the day in the format 0[1-9], 1[0-9],2[0-9] and 3[0-1] followed by "/ or ."   
    ## third select the year as a 4 digit of the date, ie year YYYY in the end. 
    
    #match=re.findall("(\d){4}-(0[1-9]|1[0-2])-(0[1-9]|1\d|2\d|3[0-1]?[0-9])",line)
    ## 
    regex=re.compile(r"^(1[0-2]|0[1-9]|[1-9])(/|\.)(0[1-9]|1\d|2\d|3[01]|[1-9])(/|\.)(\d{4})(\s|,|.)$",re.IGNORECASE)
    matches=regex.findall(line)
    if(matches):
        s=''.join(matches[0])
        fout.write((line.strip("\n")).strip() +"-- "+re.sub(regex,non_iso_to_iso(s,0),line))
        #iso_dt=non_iso_to_iso(s)
                    
print ("---------------------------------------------------------------------")


####################### Part B #####################################
## For Finding dates having DD/MM/YYYY or DD.MM.YYYY format 

## find the lines from the files. 
## first open the file and loop through the lines 

print ("All the dates which matches in the format of DD/MM/YYYY or DD.MM.YYYY")
print ("---------------------------------------------------------------------")

## first  select the day in the format 0[1-9], 1[0-9],2[0-9] and 3[0-1] followed by "/ or ."    
## second select the month in 2 digit followed by "/ or ." used a 0(1-9) & 1(0-2) 
## third  select the year as a 4 digit of the date, ie year YYYY in the end. 

regex=re.compile("^(0[1-9]|1\d|2\d|3[01]|\d)(/|\.)(1[0-2]|0[1-9]|\d)(/|\.)(\d{4})(\s|,)",re.IGNORECASE)
## Open the file and match all the patterns.
for line in (open('date.txt','r')):
    #print (line)    
    matches=regex.findall(line)
    if(matches):
        s=''.join(matches[0])
        print(s)
        fout.write((line.strip("\n")).strip() +" -- "+re.sub(regex,non_iso_to_iso(s,1),line))
        #iso_dt=non_iso_to_iso(s)

print ("---------------------------------------------------------------------")

####################### Part C #####################################
## For Finding dates having Month Day, YYYY format 

## find the lines from the files. 
## first open the file and loop through the lines 

print ("All the dates which matches in the format of DD/MM/YYYY or DD.MM.YYYY")
print ("---------------------------------------------------------------------")
for line in (open('date.txt','r')):
    #print (line)

    ## first  select the day in the format 0[1-9], 1[0-9],2[0-9] and 3[0-1] followed by "/ or ."    
    ## second select the month in 2 digit followed by "/ or ." used a 0(1-9) & 1(0-2) 
    ## third  select the year as a 4 digit of the date, ie year YYYY in the end. 
    regex=re.compile("^(JANUARY|FEBRUARY|MARCH|APRIL|MAY|JUNE|JULY|AUGUST|SEPTEMBER|NOVEMBER|DECEMBER)(\s)(0[1-9]|1\d|2\d|3[01]|\d)(,\s)(\d{4})(\s|,)$",re.IGNORECASE)
    ## 
    matches=regex.findall(line)
    if matches:
        #print(matches)
        s=''.join(matches[0])
        print(s)
        fout.write((line.strip("\n")).strip() +" -- "+ re.sub(regex,non_iso_to_iso(s,0),line))
        #iso_dt=non_iso_to_iso(s,1)
        #print(iso_dt)

print ("---------------------------------------------------------------------")
```
